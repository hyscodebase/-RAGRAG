LangChain 1.x부터 `text_splitters`, `openai`, `community` 등이 전부 **별도 패키지**로 분리되어서, `langchain.xxx` 대신 `langchain_xxx` 형태로 import해야 합니다.

| 구버전 (v0.x) | 신버전 (v1.x) |
|---|---|
| `from langchain.text_splitter import ...` | `from langchain_text_splitters import ...` |
| `from langchain.chat_models import ChatOpenAI` | `from langchain_openai import ChatOpenAI` |
| `from langchain.embeddings import OpenAIEmbeddings` | `from langchain_openai import OpenAIEmbeddings` |
| `from langchain.vectorstores import Chroma` | `from langchain_chroma import Chroma` |

````python
#다른 교재나 강의에서 langchin.text_splitter <- 이런거 그대로 쓰면 모델 없다고 오류 나는거예요
chunk_size=10
chunk_overlap=5

r_splitter = RecursiveCharacterTextSplitter(
    chunk_size = chunk_size,
    chunk_overlap = chunk_overlap
)

c_splitter = CharacterTextSplitter(
    chunk_size = chunk_size,
    chunk_overlap = chunk_overlap
)
````

````python
text1 = '청킹을 위한 예제 데이터입니다. 어떻게 쪼개지는지 보자구요!'
r_splitter.split_text(text1)
````

````python
some_text = 'Artificial intelligence has transformed the way we interact with technology in our daily lives. From voice assistants that understand natural language to recommendation systems that predict our preferences, AI is embedded in nearly every digital experience. Machine learning, a subset of AI, enables computers to learn from data without being explicitly programmed. Deep learning, which uses neural networks with multiple layers, has achieved remarkable breakthroughs in image recognition, natural language processing, and game playing. However, these advances also raise important questions about privacy, bias, and the ethical use of automated decision-making systems. As AI continues to evolve, researchers are exploring ways to make models more transparent, fair, and accountable. The field of explainable AI aims to create systems whose decisions can be understood and trusted by humans. Meanwhile, reinforcement learning is pushing the boundaries of what machines can achieve in complex environments, from robotics to autonomous driving. The future of AI promises even greater integration into healthcare, education, and scientific discovery, but it will require careful governance and collaboration between technologists, policymakers, and society at large.'
````

````python
print('오늘도')
print('\n어김없이')
print('\n\n공부를')
print('\시작한다')
````

````python
c_splitter = CharacterTextSplitter(
    chunk_size=450,
    chunk_overlap=0,
    separator=' '
)
c_splitter.split_text(some_text)
````

````python
r_splitter = RecursiveCharacterTextSplitter(
    chunk_size=450,
    chunk_overlap=0,
    separators=['\n\n', '\n', ' ', '']
)

r_splitter.split_text(some_text)
````

RecursiveCharacterTextSplitter의 경우,
가장 먼저 \n\n으로 표시된 2줄 띄어쓰기 부분을 먼저 나누고, 이때 문자 역시 공백을 기준으로 나뉜 것을 알 수 있음
마직막으로 character로 분리하기 때문에 더 세밀하게 chunking이 됨

````python
from langchain_text_splitters import TokenTextSplitter

text_splitter = TokenTextSplitter(chunk_size=1, chunk_overlap=0)

text1="Border collies are vvvery intelligent, capable of learning well, and have a lot of energy, so they need a lot of energy, so they need a lot of exercise"
text_splitter.split_text(text1)
````

````python
#tiktoken을 이용하는 방법도 있음!
from langchain_text_splitters import CharacterTextSplitter

chunk_overlap=5
chunk_size=20

splitter = CharacterTextSplitter.from_tiktoken_encoder(
    chunk_size=chunk_size,
    chunk_overlap=chunk_overlap,
    model_name = "gpt-3.5-turbo"
)
````

````python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("MIT.pdf")

#loader.load()를 호출하여 문서를 로드
pages = loader.load()

print(len(pages))
print(pages[0].page_content[0:500])
````

````python
from langchain_community.document_loaders import WebBaseLoader

# url에서 가져오기
loader = WebBaseLoader("https://ko.darktrace.com/cyber-ai-glossary/content-filtering") #임의의 url 입력

docs = loader.load()
print(docs[0].page_content[:500])
````

````python
# PDF 가져오기
loader = PyPDFLoader("MIT.pdf")

docs = []
docs.extend(loader.load())
````

````python
doc=splitter.split_documents(docs)
````

````python
#문서의 분할된 청크의 크기와 문서의 내용을 각각 출력
print(len(doc[0].page_content))
doc[0]
````
